```python
##Linear Search Answers
# Linear Search

# Target student ID to search for
target_student_id = 123456

# List of student IDs
student_ids = [112233, 445566, 789012, 123456, 654321]

# Function to perform linear search
def linearSearch(target, student_list):
    for student_id in student_list:
        if student_id == target:
            return True
    return False

# Check if the target student ID is found
if linearSearch(target_student_id, student_ids):
    print("Student Found")
else:
    print("Student Not Found")

##Strings
# Linear Search Example with Strings

# List of names
names = ["Alice", "Bob", "Charlie", "David", "Emma"]

# Function to perform linear search
def linearSearch(target, name_list):
    for name in name_list:
        if name.lower() == target.lower():
            return True
    return False

# Get input from the user
target_name = input("Enter the name to search for: ")

# Check if the target name is found
if linearSearch(target_name, names):
    print("Name Found")
else:
    print("Name Not Found")
```
